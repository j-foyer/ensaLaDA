```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# GLOBAL VARS
# number of documents
D = 50
# min and max number of words in a document
w_min = 100
w_max = 1000
# number of topics
K = 10
# number of words 
W = 100
# alpha parameter
alpha = 0.1 * np.ones(K)
# beta parameter
beta  = 0.1 * np.ones(W)
# set wordpath
WORDPATH = "words_alpha.txt"
```

```python
# load words

with open(WORDPATH,"r+") as fn:
    wdict = fn.readlines()
wdict = np.array([w.rstrip("\n") for w in wdict])
# downsample to only use W words, for speed 
wdict = np.random.choice(wdict, replace= False, size=W)
```

```python
# set number of words per document
N  = np.random.randint(low = w_min,high = w_max,size = D)
```

```python
# get theta values, dim : D x K
theta = np.random.dirichlet(alpha, size = D)
print("size of theta is : {}".format(theta.shape))
# get phi values, dim : K x W
phi = np.random.dirichlet(beta,size = K)
print("size of phi is : {}".format(phi.shape))
```

```python
# construct documents
categ = lambda p : np.argmax(np.random.multinomial(1,pvals = p))
# to keep actual words of each document
corpus_words = list()
# a count matrix of words
corpus_counts = np.zeros((D,W))
# iterate over documents
for d in range(D):
    corpus_words.append([])
    # iterate over words in each document
    for _ in range(N[d]):
        z = categ(theta[d,:])
        w = categ(phi[z,:])
        corpus_words[d].append(wdict[w])
        corpus_counts[d,w] += 1

corpus_counts = np.array(corpus_counts)
```

```python
#load
from sklearn.decomposition import LatentDirichletAllocation as LDA
```

```python
lda = LDA(n_components=K,doc_topic_prior=0.1)
lda_res = lda.fit_transform(corpus_counts)
```

```python
lda.components_.shape
```

```python
import lap
from scipy.spatial.distance import cdist
# use linear assingment problem to match word-topic distribution vectors
cost = cdist(lda.components_,phi)
_,row,col = lap.lapjv(cost)
# plot resutls
fig,ax = plt.subplots(1,2, figsize = (20,10),facecolor = "white")
ax[0].imshow(lda.components_)
ax[0].set_aspect("auto")
ax[1].imshow(phi[row,:])
ax[1].axis("auto")
ax[0].set_title(r"Estimated values of $\phi$", fontsize = 20)
ax[1].set_title(r"True values of $\phi$",fontsize = 20)

for axx in ax:
    axx.set_yticks(np.arange(10))
    axx.set_yticklabels(["Topic {}".format(k) for k in range(10)],fontsize = 15)
    axx.set_xlabel("Word no.",fontsize = 15)

plt.show()

```
